```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Task 2
f= "/Users/lengocynh/Downloads/hudata_Billionaires.listed.Forbes - data_Billionaires.listed.Forbes.csv"
df=pd.read_csv(f)
print(df.head(10))
```

```python
print(df.describe())
```

```python
#task 3
# Sắp xếp DataFrame theo một cột cụ thể (ví dụ: cột 'ColumnName' từ cao đến thấp)
sorted_df = df.sort_values(by='Net_Worth_($B)', ascending=False)
print(sorted_df)
```

```python
output_csv_path = '/Users/lengochuynh/Downloads/task3_1.csv'
# Lưu DataFrame đã sắp xếp vào một tệp CSV mới
sorted_df.to_csv(output_csv_path, index=False)
```

```python
#Task 3
# Lọc và sắp xếp theo quốc gia làm việc
top_countries = df.groupby('Country_Working_In').size().sort_values(ascending=False).head(10)
# In ra top 10 quốc gia
print(top_countries)
# Task 4
#vẽ biểu đồ line chart
plt.figure(figsize=(10, 6))
plt.plot(top_countries.index, top_countries.values, marker='o', label='Number of Billionaires', color='blue')
plt.title('Top 10 Countries with Most Billionaires')
plt.xlabel('Country')
plt.ylabel('Number of Billionaires')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
# vẽ biểu đồ pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Billionaires Across Top 10 Countries')
plt.show()
```

```python
#Task 3
top_citizenships = df.groupby('Citizenship').size().sort_values(ascending=False).head(10)

# In ra top 10 quốc tịch
print(top_citizenships)
#Task 4
#vẽ biểu đồ line chart
plt.figure(figsize=(10, 6))
plt.plot(top_citizenships.index, top_citizenships.values, marker='o', label='Number of Billionaires', color='green')
plt.title('Top 10 Countries with Most Billionaires Based on Citizenship')
plt.xlabel('Country')
plt.ylabel('Number of Billionaires')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
#Task 4
# vẽ biểu đồ pie chart
# Vẽ Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(top_citizenships, labels=top_citizenships.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Billionaires Across Top 10 Countries Based on Citizenship')
plt.show()
```

```python
#Task 3
# Lọc và sắp xếp theo độ tuổi
top_ages = df.groupby('Age').size().sort_values(ascending=False).head(5)

# In ra top 5 độ tuổi
print(top_ages)
#Task 4
#vẽ biểu đồ lien chart
plt.figure(figsize=(10, 6))
plt.plot(top_ages.index, top_ages.values, marker='o', label='Number of Billionaires', color='purple')
plt.title('Top 5 Ages with Most Billionaires')
plt.xlabel('Age')
plt.ylabel('Number of Billionaires')
plt.legend()
plt.grid(True)
plt.show()
#vẽ biểu đồ pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_ages, labels=top_ages.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Ages with Most Billionaires')
plt.show()
```
